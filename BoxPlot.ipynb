```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'Data Analysis - Caselet#2.xlsx'
sheet_name = 'DifferentFormat'

# Read the Excel file into a DataFrame
df = pd.read_excel(file_path,sheet_name = sheet_name)

# Create a box plot for TPT values for each Team Member
plt.figure(figsize=(12, 6))
df.boxplot(column='TPT', by='Team Member', grid=False)
plt.title('Boxplot for Team Members and TPT')
plt.xlabel('Team Member')
plt.ylabel('TPT')
plt.xticks(rotation = 90)
 
plt.show()
```
